```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import nltk
```

```python
locs = pd.read_csv('Review_db.csv')
locs
```

```python
rating_count =locs.groupby('Place').count()['Rating'].reset_index()
rating_count.rename(columns={'Rating':'No. of rating'},inplace=True)
rating_count
```

```python
avg_rating=pd.DataFrame(locs.groupby('Place')['Rating'].mean())
avg_rating
```

```python
locs.drop(columns =['Date','Raw_Review','Name','Rating'], inplace = True)
```

```python
locs=locs.merge(rating_count, on='Place')
locs
```

```python
locs=locs.merge(avg_rating, on='Place')
locs
```

```python
locs
```

```python
pop_loc=locs[locs['No. of rating']>500].sort_values('Rating',ascending=False)
pop_loc
```

```python
pop_loc=pop_loc.drop_duplicates()
pop_loc
```

```python
places = pop_loc['Place'].unique()
place_review=pd.DataFrame(columns=['Place','Review'])
place_review['Place'] = places
reviews=[]
for place in places:
    st =""
    for item in pop_loc[pop_loc['Place']==place]['Review']:
        st+=item
    reviews.append(st)
place_review['Review'] = reviews
```

```python
place_review
```

```python
pop_loc.drop(columns =['Review'], inplace = True)
pop_loc
```

```python
pop_loc = pop_loc.drop_duplicates()
```

```python
pop_loc.set_index('Place',inplace = True)
place_review.set_index('Place',inplace=True)
pop_loc['Reviews']=""*pop_loc.shape[0]
for place in pop_loc.index:
    pop_loc.loc[place,'Reviews']=place_review.loc[place,'Review']
pop_loc
```

```python
pop_loc.reset_index(inplace=True)
pop_loc
```

# Recommend Popular Cities using Collaborating Filter


```python
pt = pop_loc.pivot_table(index ='City',columns='Place',values='Rating')
pt.fillna(0,inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
def recommend_city(city):
    index = np.where(pt.index==city)[0][0]
    popular_cities = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:11]

    for i in popular_cities:
        print(pt.index[i[0]])
```

```python
recommend_city('Vrindavan')
```

# Place Recommendation using Content based Filtering

```python
pop_loc
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
pop_loc['Reviews']=pop_loc['Reviews'].apply(stem)
```

```python
pop_loc
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 10000, stop_words='english')
```

```python
vectors = cv.fit_transform(pop_loc['Reviews']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
def recommend_place(place):
    place_index = pop_loc[pop_loc['Place']==place].index[0]
    distances = similarity[place_index]
    similar_places = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:11]
    for i in similar_places:
        print(pop_loc.iloc[i[0]].Place + " - " + pop_loc.iloc[i[0]].City)
```

```python
recommend_place("Science City")
```
